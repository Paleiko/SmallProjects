```python
mark = {1: "x", -1: "o", 0: " "}
players = {1: "1", -1: "2"}

def start():
    running = False
    play = input("Want to start a game? (y/n)")
    if play == "y":
        return True
    elif play == "n":
        print("Ok, let's play another time.")
        return False
    else:
        print("I didn't understand that.")
        start()

def print_horiz_line():
    print(" ---" * board_size)

def print_vert_line(row = [0]*3):
    print("| {} | {} | {} |".format(mark[row[0]], mark[row[1]], mark[row[2]]))

def draw_grid(grid, count):
    print("---------------------------")
    for i in range(len(grid)):
        print_horiz_line()
        print_vert_line(grid[i])
    print_horiz_line()
    print("---------------------------")
    return count+1    

def test_win(grid):
    max_v = len(grid)
    rows = [abs(sum(a)) for a in grid]
    columns = [abs(sum(b)) for b in zip(*grid)]
    d1 = abs(sum(grid[i][i] for i in range(max_v)))
    d2 = abs(sum(grid[i][max_v - 1 - i] for i in range(max_v)))

    if max_v in [*rows,*columns,d1,d2]:
        return True
    
def pick_a_field():
    return input("Pick an index like 1,1 or 3,3: ").split(",")
    
def switch_player(num):
    if num == 1: return -1
    else: return 1
    
def update_grid(grid):
    x,y = map(int,pick_a_field())
    time.sleep(0.2)
    if grid[x-1][y-1] == 0:
        grid[x-1][y-1] = player
        clear_output()
        return grid
    else:
        print("Sorry that field is already used. Choose again.")
        update_grid(grid)
```
